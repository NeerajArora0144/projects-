```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')

# setting canvas and dpi
plt.figure(figsize = (8,5),dpi=300)

# Title 
plt.title('Gas Prices over Time (in USD)',fontdict={'fontweight':'bold', 'fontsize' : '18'})

## For selecting specific Countries one by one
# plt.plot(gas.Year, gas.USA,'b.-' ,label = 'United States')
# plt.plot(gas.Year, gas.Canada,'r.-', label = 'Canada - mini punjab')
# plt.plot(gas.Year, gas['South Korea'],'g.-', label = 'South Korea')
# plt.plot(gas.Year, gas.Australia,'y.-', label = 'Australia')


## For selecting all Countries
# for country in gas:
#     if country != 'Year':
#         print(country)
#         plt.plot(gas.Year,gas[country],label=country, marker='.')


## For selecting specific Countries from list
countries_to_look_at = ['Australia','USA','Canada','South Korea']
for country in countries_to_look_at:
    if country in countries_to_look_at:
        plt.plot(gas.Year,gas[country],label=country, marker='.')


# adding legend
plt.legend(bbox_to_anchor=(1,1))
# plt.legend(loc='upper left', bbox_to_anchor=(1,1))


# adding X label
plt.xticks(gas.Year[::3])



# adding X label
plt.xlabel('Years')
# adding Y label
plt.ylabel('Price')

# saving the graph with specific dpi
plt.savefig('Gas_Price_figure.png', dpi=300)

# printing output
plt.show()
```
